### Spatial autocorr

We take the autocorrelation plot for the temperature, with the upscaling dataset, the original dataset and the different downscaled datasets

```python
ds = metrics.corr(ups, 200, 'T_2M', 'rlon', 12)
```

```python
normal = metrics.corr(week_2km, 800, 'T_2M', 'rlon')
```

```python
test = []
for ds in methods :
    test.append(metrics.corr(ds, 800, 'T_2M', 'rlon'))
```

We then plot all those autocorrelations

```python
sns.lineplot(data = ds, x = 'lag', y = 'corr')
sns.lineplot(data = normal, x = 'lag', y = 'corr')
for i in test :
    sns.lineplot(data = i, x = 'lag', y = 'corr')
```

It is interesting to see what happens for the small lags, so we keep only the first values

```python
ds_bis = ds[ds['lag'] <= 24]
normal_bis = normal[normal['lag'] <= 24]
test_bis = []
for bis in test :
    test_bis.append(bis[bis['lag'] <= 24])
```

```python
ds_bis
```

```python
sns.lineplot(data = ds_bis, x = 'lag', y = 'corr', label = 'coarse resolution')
sns.lineplot(data = normal_bis, x = 'lag', y = 'corr', label = 'og data')
for i in test_bis :
    sns.lineplot(data = i, x = 'lag', y = 'corr')
```

We the that the downscaled methods are following the trend of the upscale dataset.
